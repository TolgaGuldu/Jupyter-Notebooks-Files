#### Use the lecture code and introduction below to use sklearn to fit a linear model to the data generated below. 
- In the example below I show the steps of applying a two dimensional linear plane fit. 
- Input is two dimensional  X=(xx,yy)-> t 
- Output is one dimensional

#### Bonus HW

- Follow the example steps to generate three dimensional input grid
- Input is two dimensional  (xx,yy,zz)->t
- Output is one dimensional





```python
import numpy as np
np.random.seed(0)
x = np.linspace(-1,1,20)
y = np.linspace(-2, 2, 20)
xx, yy = np.meshgrid(x, y)
num_points = np.prod(xx.shape)

print("number of points:",num_points)
def f(x,y):
    t = -3*x + 4*y -5 + np.random.rand(1)
    return t
# use zip(xx,yy) for all points in the grid

targets = f(xx,yy)
print(targets.shape)
print("targets:",targets)
```

1) First create a matrix X which has two columns formed of xx and yys.


```python
X = np.zeros(shape=(num_points,2))
print(X.shape)
X[:,0] = xx.flatten()
X[:,1] = yy.flatten()

# complete the rest

y = np.array(targets).reshape(num_points,1)
print(y.shape)
```

```python
print("X:",X)
```

```python
print("y:",y)
```

2) Create a linear model from sklearn

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression(fit_intercept=True)

model.fit(X,y)
print("a,b:", model.coef_)
print("c:, ",model.intercept_)
```

print the fitted model coefficients and intercept

```python
x = np.linspace(-1,1,20)
y = np.linspace(-2, 2, 20)
z = np.linspace(-1.5, 1.5, 10)
xx, yy,zz = np.meshgrid(x, y, z)
num_points = np.prod(xx.shape)

print("number of points:",num_points)
def f(x,y,z):
    t = 2*x + y -10*z + 50+ np.random.rand(1)
    return t
# use zip(xx,yy) for all points in the grid

targets = f(xx,yy,zz)
print(targets.shape)
```

```python
X = np.zeros(shape=(num_points,3))
print(X.shape)
X[:,0] = xx.flatten()
X[:,1] = yy.flatten()
X[:,2] = zz.flatten()

y = targets.flatten()
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression(fit_intercept=True)

model.fit(X,y)
print("a,b,c:", model.coef_)
print("d:, ",model.intercept_)
```
