#### 1b) Remove the data  which does not have mileage information (like numpy you use boolean indexing)

#### 1d) What is the mean fuel mileage for Automatic cars? 
 What is the mean fuel mileage for Manual Hatchbacks?
 
- Be carefull that some attributes have spaces in the names. cars2.GearType==' Automatic'

```python
cars_auto=cars2[cars2.GearType==' Automatic'].copy().reset_index()
print("Mean automatic: ",cars_auto.Mileage.mean())
cars_man=cars2[cars2.GearType==' Manual'].copy().reset_index()
print("Mean manual: ",cars_man.Mileage.mean())
```

#### 1e) What is the mean fuel mileage for TypeSedan cars?
 What is the mean fuel mileage for Type Hatchbacks?

```python
sedans = cars2[cars2.Type=='Sedan']
sedans.head(10)
```

```python
cars_sedan=cars2[cars2.Type=='Sedan'].copy().reset_index()
print("Mean sedan: ",cars_sedan.Mileage.mean())
cars_hb=cars2[cars2.Type=='Hatchback'].copy().reset_index()
print("Mean hb: ",cars_hb.Mileage.mean())
```

#### 1f) 1. Hypothesis :  Automatic cars mileage is lower than Manual cars. Do the t-tests on data to show if the mean difference is significant

```python
stats.ttest_ind(cars_auto.Mileage, cars_man.Mileage, equal_var=True)
```

```python
# null hypothesis is that mean differerence is not significant. 
print("ttest pvalue is near 0 (much less than 0.05), So the null hypothesis is rejected. Mean difference is significant!")
```

#### 1g) 2. Hypothesis :  Sedan cars mileage is lower than Hb cars. Do the t-tests on data to show if the mean difference is significant

```python
stats.ttest_ind(cars_sedan.Mileage, cars_hb.Mileage, equal_var=True)
```

```python
# null hypothesis is that mean differerence is not significant. 
print("Again ttest pvalue is near 0 (much less than 0.05), So the null hypothesis is rejected. Mean difference is significant!")
```

#### 1h) Above t-tests assume normal distribution of sample. 
Does automatic car mileage have normal distribution?
Does manual car mileage have normal distribution?

Does sedan car mileage have normal distribution?
Does hb car mileage have normal distribution?

```python
print(stats.shapiro(cars_man.Mileage))
print(stats.shapiro(cars_auto.Mileage))
print(stats.shapiro(cars_sedan.Mileage))
print(stats.shapiro(cars_hb.Mileage))
```

```python
# null hypothesis is data from normal distribution. 
print('Again pvalues are less than 0.05, So the null hypothesis is rejected',
      ' none of the data seem to be from normal distribution!')
```
