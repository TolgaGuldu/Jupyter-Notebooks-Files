```python
### or create a random dataset with numpy
rng = np.random.RandomState(42)

x = 10 * rng.rand(50)
print(x)
y = 2 *x - 1 + 0.1*rng.rand(50)
print(y)

plt.plot(x,y,'o')
```

#### Now lets learn to fit a linear  model
https://scikit-learn.org/stable/modules/linear_model.html

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()

reg.fit([[0], [1], [2]], [0, 0.2, 0.4])


print(reg)
print(reg.coef_)
print(reg.intercept_)

reg.predict([[3],[-1]])


```

```python
# https://scikit-learn.org/stable/modules/linear_model.html
from sklearn.linear_model import LinearRegression

model = LinearRegression(fit_intercept=True)
x1 = x[:,np.newaxis]
model.fit(x1,y)
print(model.coef_)
print(model.intercept_)

```

```python
tstx = np.linspace(-1,10,20)
tstxmat = tstx[:, np.newaxis]
tsty = model.predict(tstxmat)

plt.plot(x,y,'ro')
plt.plot(tstx, tsty, 'o')
```
