## Q4 (10pts)

Using only python syntax, numpy or pandas , stats, matplotlib libs

We are trying to compare two classifiers. You have made a cross validation test and accuracy scores are as below

A: [ 61.66, 61.72, 60.86, 61.23, 61.28]
B: [ 62.03, 61.50, 60.43, 61.20, 61.19]

Which one is better? Prove your hypothesis with statistical t-tests.

```python
# a) 
import numpy as np
from numpy import logical_and as land

A = np.array([61.66, 61.72, 60.86, 61.23, 61.28])
B = np.array([62.03, 61.50, 60.43, 61.20, 61.19])
```

```python
#assume pos = 1 neg = 0 
tpos = np.count_nonzero(land(B == 1, A==1)+61) # label is pos , prediction is pos
fneg = np.count_nonzero(land(B == 0, A==0)+61) # label is pos, prediction is neg
fpos = np.count_nonzero(land(B == 1, A==0)+61)  # label is neg, prediction is pos 
recall = tpos/(tpos+fneg)

precision = tpos/(tpos+fpos)

print("recall",recall)
print("precision", precision)
f_1 =2/((1/precision)+(1/recall))
print("f1 score", f_1)
```

## Q7 (10pts)

For a binary classification true labels and predictions are given below. You can use sklearn library functions.

labels = [1,2,1,2,1,2,1,1,1,1,2,2,2,1,1]

predictions = [1,1,2,2,1,2,2,2,1,1,2,2,2,1,2]

Assume that label 2 is for positive, 1 is for negative class.

Calculate and print 

a) number of true positives
tp,
and number of false positives
fp .

b) confusion matrix (rows show true labels, columns show prediction results)
tn|fp
fn|tp

c) recall, precision, f-1 scores


```python
# a) 
import numpy as np
from numpy import logical_and as land

labels = [1,2,1,2,1,2,1,1,1,1,2,2,2,1,1]
predictions = [1,1,2,2,1,2,2,2,1,1,2,2,2,1,2]
```

```python
#assume pos = 2 neg = 1 
tpos = np.count_nonzero(land(predictions == 2, labels==2)+1) # label is pos , prediction is pos
fneg = np.count_nonzero(land(predictions == 1, labels==1)+1) # label is pos, prediction is neg
fpos = np.count_nonzero(land(predictions == 1, labels==2)+1)  # label is neg, prediction is pos 
recall = tpos/(tpos+fneg)

precision = tpos/(tpos+fpos)

print("recall",recall)
print("precision", precision)
f_1 =2/((1/precision)+(1/recall))
print("f1 score", f_1)
```
