## q1


Using no python libraries. 

 a) Define an array of n elements. 
 
Using python syntax calculate 

 b) sum of the elements
 
 c) mean of the elements
 
 d) std dev of the elements
   



```python
array = [5,13,26,53,32,53,67,32,65,42,145,32,54,321]
print(sum(array))
```

```python
mean = sum(array)/len(array)
print(mean)
```

```python
std = (sum([(mean-x)**2 for x in array])/(len(array)-1))**(1.0/2)
print(std)
```

## q4

Using only python syntax, numpy or pandas,matplotlib lib

a) Read image.png into a numpy array

b) calculate sum, mean, std, max, min of  red channel

c) plot histogram blue channel

d) plot scatter plot of red vs greens.

e) create and display a binary mask where object is foreground and the rest is background

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

image = mpimg.imread('dog.jpg')

plt.imshow(image)
```

```python
# Separate im into 3 channels.
im_r = image[:,:,0]
im_g = image[:,:,1]
im_b = image[:,:,2]

plt.figure(figsize=(30,10))
plt.imshow(im_r)
plt.colorbar()
plt.title('red')

print(im_r.sum())
print(im_r.mean())
print(im_r.std())
print(im_r.max())
print(im_r.min())
```

```python
plt.hist(im_b.flatten(), 100)
```

```python
plt.scatter(im_r.flatten(),im_g.flatten(),marker = 'o')
plt.show()
```
