```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
x=["1.ABD","2.ÇİN","3.JAPONYA","4.ALMANYA","5.İNGİLTERE","6.FRANSA"]
y=[20410,14090,5170,4210,2940,2930]
```

```python
plt.plot(x,y,"black")
plt.xlabel("ÜLKELER")
plt.ylabel("MİLLİ GELİR")
plt.title("ülkelere göre milli gelir")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"r--") 
plt.subplot(1,2,2)
z=[63039,9957,40649,51158,44162,44915]
plt.plot(x,z,"y*-")  
plt.tight_layout()
```

```python
a = plt.figure()
axes=a.add_axes([0.1,0.1,0.9,0.9])
axes.plot(x,y)
axes.set_xlabel("ülkeler")
axes.set_ylabel("miili gelir")
axes.set_title("ülkeler göre mili gelir")
```

```python
f=plt.figure()
axes1=f.add_axes([0.1,0.1,0.9,0.9])
axes2=f.add_axes([0.5,0.5,0.4,0.4])
axes1.plot(x,y,"y")
axes2.plot(x,z,"r")
```

```python
fig,axes=plt.subplots(1,2)

axes[0].plot(x,y,"b")
axes[0].set_title("mili gelir")

axes[1].plot(x,z,"g")
axes[1].set_title("kişi başına")

plt.tight_layout()
```

```python
a = plt.figure()
axes=a.add_axes([0.1,0.1,0.9,0.9])
axes.plot(x,y,color="red",linewidth=1,linestyle="-",marker="*",markersize=50,markerfacecolor="yellow",markeredgewidth=4,markeredgecolor="blue",alpha=0.5)
axes.set_xlabel("ülkeler")
axes.set_ylabel("miili gelir")
axes.set_title("ülkeler göre mili gelir")

axes.set_xlim([2,6])
axes.set_ylim([0,7500])
```

```python
a = plt.figure()
axes=a.add_axes([0.1,0.1,1,1])
axes.plot(x,y,color="red")
axes.set_xlabel("ülkeler")
axes.set_ylabel("miili gelir")
axes.set_title("ülkeler göre mili gelir")

axes.set_xlim([2,6])
axes.set_ylim([0,7500])
```

# pie charts

```python
labels = 'TÜRKİYE', 'ABD', 'ÇİN', 'ALMANYA'
sizes = [15, 35, 35, 15]
explode = (0.2, 0, 0, 0.2)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
```
